# 一、自相关函数 
Auto Correlation Function (ACF)
## 1.1 定义与计算

理论：**自相关函数$\rho_k$**
$$
\begin{align*}
&\text{均值函数：}\mu _t=E\left( Y_t \right) , t=\text{0,}\pm \text{1,}\pm \text{2,}...
\\
&\text{自协方差函数：}\gamma _{t,s}=Cov\left( Y_t,Y_s \right) =E\left[ \left( Y_t-\mu _t \right) \left( Y_s-\mu _s \right) \right] 
\\
&\text{自相关函数：}\rho _{t,s}=Corr\left( Y_t,Y_s \right) =\frac{Cov\left( Y_t,Y_s \right)}{\sqrt{Var\left( Y_t \right) Var\left( Y_s \right)}}=\frac{\gamma _{t,s}}{\sqrt{\gamma _{t,t}\gamma _{s,s}}}
\\
&\text{（宽）平稳时：}
\\
&自相关函数：\rho _k=\frac{\gamma _k}{\gamma _0}
\end{align*}
$$

估计：**样本自相关函数 $r_k$**
$$
r_k=\frac{\sum_{t=k+1}^n{\left( Y_t-\bar{Y} \right) \left( Y_{t-k}-\bar{Y} \right)}}{\sum_{t=1}^n{\left( Y_t-\bar{Y} \right) ^2}}, k=\text{1,}2\cdots  （k=0时显然为1不考虑）
$$

## 1.2 MA(q)模型

### $\rho_k$模式特点
滞后q阶后截尾，$\rho_k=0, k>q$

### $r_k$模式特点
当**k>q**时，用$$\pm 2\sqrt{\frac{1}{n}\left[ 1+2\sum_{j=1}^q{r_{j}^{2}} \right]}$$作为$r_k$的临界极限值来检验MA(q)模型是正确的零假设。

#### 理论支撑：样本自相关函数的大样本结论（6.1.2）（6.1.9~6.1.11）

假设$Y_t=\mu +\sum_{j=0}^{\infty}{\psi _je_{t-j}}$，其中$e_t$独立同分布并具有零均值和有限非零同方差，进一步假设
$$
\sum_{j=0}^{\infty}{|\psi _j|<\infty},\sum_{j=0}^{\infty}{j\psi _j^2<\infty}
$$（任意平稳ARMA模型都满足这一假设），

那么对于任意固定的m，当n趋于无穷时，联合分布
$$
\sqrt{n}\left( r_1-\rho _1 \right) ,\sqrt{n}\left( r_2-\rho _2 \right) ,\cdots ,\sqrt{n}\left( r_m-\rho _m \right)
$$
逼近均值为0、方差为$c_{jj}$、协方差为$c_{ij}$的联合正态分布，其中
$$
c_{ij}=\sum_{k=-\infty}^{\infty}{\left( \rho _{k+i}\rho _{k+j}+\rho _{k-i}\rho _{k+j}-2\rho _i\rho _k\rho _{k+j}-2\rho _j\rho _k\rho _{k+i}+2\rho _i\rho _j\rho _{k}^{2} \right)}
$$
换句话说，当n趋于无穷时，$(r_1,r_2,...,r_m)$的分布逼近m元联合正太分布$N\left( \left( \rho _1,\rho _2,\cdots ,\rho _m \right) ,\left[ \begin{matrix}
	\frac{c_{11}}{n}&		\cdots&		\frac{c_{1m}}{n}\\
	\vdots&		\ddots&		\vdots\\
	\frac{c_{m1}}{n}&		\cdots&		\frac{c_{mm}}{n}\\
\end{matrix} \right] \right) $

* 当n充分大时，我们将认为$r_k$近似服从$N(\rho_k,c_{kk}/n)$，且$Corr(r_k,r_j)\approx c_{ki}/\sqrt{c_{kk}c_{jj}}$。可知$r_k$的近似方差与样本的大小成反比，而当n较大时，$Corr(r_k,r_j)$近似为常数。

* 对于一般的MA(q)过程，有$$
c_{kk}=1+2\sum_{j=1}^q{\rho_{j}^{2}}, k>q\\
(i=j>q，此时c_{ij}中只有第一项可以不为0，令-q<=k+i<=q)\\
于是Var\left( r_k \right) \approx \frac{c_{kk}}{n}=\frac{1}{n}\left[ 1+2\sum_{j=1}^q{\rho _{j}^{2}} \right] , k>q
$$

# 二、偏自相关函数
Partial Auto Correlation Function (ACF)
## 2.1 定义

定义一（6.2.1）

定义二（6.2.2）

关于定义二的解释：
$$
\begin{align*}
&\min\text{ }E\left( Y_t-\beta _1Y_{t-1}-\beta _2Y_{t-2}-...-\beta _{k-1}Y_{t-k+1} \right) ^2\\
&\Rightarrow \begin{cases}
	E\left( Y_t-\beta _1Y_{t-1}-\beta _2Y_{t-2}-...-\beta _{k-1}Y_{t-k+1} \right) Y_{t-1}=0\\
	E\left( Y_t-\beta _1Y_{t-1}-\beta _2Y_{t-2}-...-\beta _{k-1}Y_{t-k+1} \right) Y_{t-2}=0\\
	\cdots\\
	E\left( Y_t-\beta _1Y_{t-1}-\beta _2Y_{t-2}-...-\beta _{k-1}Y_{t-k+1} \right) Y_{t-k+1}=0\\
\end{cases}
\\
&\Rightarrow \begin{cases}
	\gamma _1-\beta _1-\beta _2\gamma _1-...-\beta _{k-1}\gamma _{k-2}=0\\
	\gamma _2-\beta _1\gamma _1-\beta _2-...-\beta _{k-1}\gamma _{k-3}=0\\
	\cdots\\
	\gamma _{k-1}-\beta _1\gamma _{k-2}-\beta _{k-3}-...-\beta _{k-1}=0\\
\end{cases}
\\
&\Rightarrow \left[ \begin{matrix}
	1&		\gamma _1&		\cdots&		\gamma _{k-2}\\
	\gamma _1&		1&		\cdots&		\gamma _{k-3}\\
	\vdots&		\vdots&		\ddots&		\vdots\\
	\gamma _{k-2}&		\gamma _{k-3}&		\cdots&		1\\
\end{matrix} \right] \left[ \begin{array}{c}
	\beta _1\\
	\beta _2\\
	\vdots\\
	\beta _{k-1}\\
\end{array} \right] =\left[ \begin{array}{c}
	\gamma _1\\
	\gamma _2\\
	\vdots\\
	\gamma _{k-1}\\
\end{array} \right] 
\end{align*}
$$

## 2.2 计算

理论：**偏自相关系数$\phi_{kk}$**

* 对于具有自相关函数$\rho_k$的任意平稳过程，可以证明$\phi_{kk}$满足Yule-Walker方程(6.2.7) (6.2.8)：
$$
\left[ \begin{matrix}
	1&		\rho _1&		\cdots&		\rho _{k-1}\\
	\rho _1&		1&		\cdots&		\rho _{k-2}\\
	\vdots&		\vdots&		\ddots&		\vdots\\
	\rho _{k-1}&		\rho _{k-2}&		\cdots&		1\\
\end{matrix} \right] \left[ \begin{array}{c}
	\phi _{k1}\\
	\phi _{k2}\\
	\vdots\\
	\phi _{kk}\\
\end{array} \right] =\left[ \begin{array}{c}
	\rho _1\\
	\rho _2\\
	\vdots\\
	\rho _k\\
\end{array} \right] 
$$

* Levinson(1947)和Durbin(1960)给出了上述方程的递归求解方法（6.2.9）：
$$
\begin{cases}
	\phi _{kk}=\frac{\rho _k-\sum_{j=1}^{k-1}{\phi _{k-\text{1,}j}\rho _{k-j}}}{1-\sum_{j=1}^{k-1}{\phi _{k-\text{1,}j}\rho _j}}\\
	\phi _{k,j}=\phi _{k-\text{1,}j}-\phi _{kk}\phi _{k-\text{1,}k-j}, j=\text{1,2,...,}k-1\\
\end{cases}
$$

$$
\begin{align*}
\phi _{11}&=\rho _1
\\
\rightarrow \phi _{22}&=\frac{\rho _2-\phi _{11}\rho _1}{1-\phi _{11}\rho _1}, \phi _{21}=\phi _{11}-\phi _{22}\phi _{11}
\\
\rightarrow \phi _{33}&=\frac{\rho _3-\phi _{21}\rho _2-\phi _{22}\rho _1}{1-\phi _{21}\rho _1-\phi _{22}\rho _2}, \phi _{31}=\phi _{21}-\phi _{33}\phi _{22}, \phi _{32}=\phi _{22}-\phi _{33}\phi _{\begin{array}{c}
	21\\
\end{array}}
\\
\rightarrow \phi _{44}&=\frac{\rho _4-\phi _{31}\rho _3-\phi _{32}\rho _2-\phi _{33}\rho _1}{1-\phi _{31}\rho _1-\phi _{32}\rho _2-\phi _{33}\rho _3}, \phi _{41}=...
\end{align*}
$$

估计：**样本偏自相关系数$\hat{\phi}_{kk}$**

用样本自相关函数r来估计$\rho$，代入求解得到$\phi_{kk}$的估计，称为样本偏自相关函数并记为$\hat{\phi}_{kk}$。

## 2.3 AR(p)模型

### $\rho_k$模式特点

滞后p阶后截尾。

$\phi_{kk}=0, k>p$ (6.2.4)

$\phi_{pp}=\phi_p$ (6.2.8)

### $r_k$模式特点

当k>q时，用$\frac{\pm 2}{\sqrt{n}}$作为$\hat{\phi}_{kk}$的临界极限值来检验AR(p)模型是正确的零假设。

#### 理论支撑

# 三、扩展的自相关函数 EACF
如果混合ARMA模型的AR部分是已知的，则从观测时间序列中“滤出”自回归部分将得到一个纯MA过程，该过程的ACF具有截尾的特征。自回归系数可以通过有限次的一系列回归来估计。

**例子：对真实模型是ARMA(1,1)的情况**

$$
Y_{t}=\phi Y_{t-1}+e_{t}-\theta e_{t-1}
$$

结论1：此时$Y_t$对$Y_{t-1}$的简单线性回归得到$\phi$的不一致估计量。实际上，理论回归系数等于$
\rho_{1}=(\phi-\theta)(1-\phi \theta) /\left(1-2 \phi \theta+\theta^{2}\right)
$，而不是$\phi$。

结论2：第二次回归是$Y_t$对$Y_{t-1}$和第一次回归残差的一次滞后进行的多元回归，第二次回归中$Y_{t-1}$的系数用$\widetilde{\phi}$表示，证明是$\phi$的一致估计量。

证明：

结论1：

$$
\begin{align*}
Y_t&=\phi Y_{t-1}+e_t-\theta e_{t-1}
\\
估计：\min\text{ }\sum_{t=2}^T{\left( Y_t-\hat{Y}_t \right) ^2}&=\min\text{ }\sum_{t=2}^T{\left( Y_t-a Y_{t-1} \right) ^2}
\\
理论：\min\text{ }& E\left( Y_t-a Y_{t-1} \right) ^2\\
\Rightarrow &E\left( Y_tY_{t-1} \right) =a E\left( Y_{t-1}Y_{t-1} \right) \\
\Rightarrow &a =\frac{\gamma _1}{\gamma _0}=\rho _1, \text{此即简单线性回归的理论回归系数。}\\
接下来求\rho_1的表达式：
\\
\text{记}E\left( e_t^2 \right) &=\xi ^2,
\\
\gamma _0&=E\left( Y_tY_t \right) 
\\
&=E\left[ \left( \phi Y_{t-1}+e_t-\theta e_{t-1} \right) \left( \phi Y_{t-1}+e_t-\theta e_{t-1} \right) \right] 
\\
&=\phi ^2\gamma _0-\theta \phi \xi ^2+\xi ^2-\theta \phi \xi ^2+\theta ^2\xi ^2
\\
\Rightarrow \gamma _0&=\frac{\left( \theta ^2+1-2\theta \phi \right) \xi ^2}{1-\phi ^2}
\\
\gamma _1&=E\left( Y_tY_{t-1} \right) 
\\
&=E\left[ \left( \phi Y_{t-1}+e_t-\theta e_{t-1} \right) Y_{t-1} \right] 
\\
&=\phi \gamma _0-\theta \xi ^2
\\
\Rightarrow \rho _1
&=\frac{\gamma _1}{\gamma _0}\\
&=\frac{\phi \gamma _0-\theta \xi ^2}{\gamma _0}\\
&=\left( \phi \gamma _0-\theta \xi ^2 \right) \frac{1-\phi ^2}{\left( \theta ^2+1-2\theta \phi \right) \xi ^2}\\
&=\frac{\left( \phi -\theta \right) \left( 1-\phi \theta \right)}{1-2\phi \theta +\theta ^2}\\
因此，简单线性回归的理论回归&系数a=\rho_1=\frac{\left( \phi -\theta \right) \left( 1-\phi \theta \right)}{1-2\phi \theta +\theta ^2}，而不是\phi
\end{align*}
$$

结论2：

$$
\begin{align*}
\min\text{ }&E\left( Y_t-\left( a_1Y_{t-1}+a_2\left( Y_{t-1}-\rho _1Y_{t-2} \right) \right) \right) ^2
\\
\Rightarrow &\left\{ \begin{array}{c}
	\begin{array}{c}
	\begin{array}{l}
	E\left[ Y_{t-1}\left( Y_t-a_1Y_{t-1}-a_2\left( Y_{t-1}-\rho _1Y_{t-2} \right) \right) \right] =0\\
	E\left[ \left( Y_{t-1}-\rho _1Y_{t-2} \right) \left( Y_t-a_1Y_{t-1}-a_2\left( Y_{t-1}-\rho _1Y_{t-2} \right) \right) \right] =0\\
\end{array}\\
\end{array}\\
\end{array} \right. 
\\
\Rightarrow &\begin{cases}
	\gamma _1=a_1\gamma _0+a_2\left( \gamma _0-\rho _1\gamma _1 \right)\\
	\gamma _1-\rho _1\gamma _2-a_1\gamma _0+a_1\rho _1\gamma _1-a_2\left( \gamma _0-\rho _1\gamma _1 \right) +\rho _1a_2\left( \gamma _1-\rho _1\gamma _0 \right) =0\\
\end{cases}
\\
\Rightarrow &\begin{cases}
	a_1+\left( 1-\rho _{1}^{2} \right) a_2=\rho _1\\
	\left( \rho _{1}^{2}-1 \right) a_1+\left( \rho _{1}^{2}-1 \right) a_2=\rho _2\rho _1-\rho _1\\
\end{cases}
\\
\Rightarrow &\begin{cases}
	a_1=\frac{\rho _2}{\rho _1}\\
	a_2=\frac{\rho _{1}^{2}-\rho _2}{\rho _1\left( 1-\rho _{1}^{2} \right)}\\
\end{cases}
\\
\rho _1&=\phi -\frac{\theta \xi ^2}{\gamma _0}
\\
\text{类似地得到}\rho _2&=\phi \left( \phi -\frac{\theta \xi ^2}{\gamma _0} \right)
\\
\text{于是}a_1&=\frac{\rho _2}{\rho _1}=\phi \text{。}
\\
\text{或者这样得到}\frac{\rho _2}{\rho _1}=\phi\text{：}
\\
Y_t&=\phi Y_{t-1}+e_t-\theta e_{t-1}
\\
\text{当}k>\text{1时，}&E\left[ \left( e_t-\theta e_{t-1} \right) Y_{t-k} \right] =\text{0。}
\\
\text{于是}E\left[ Y_tY_{t-k} \right] &=\phi E\left[ Y_{t-1}Y_{t-k} \right] +E\left[ \left( e_t-\theta e_{t-1} \right) Y_{t-k} \right] 
\\
\Rightarrow \gamma _k&=\phi \gamma _{k-1}, k>1
\\
\Rightarrow \frac{\rho _k}{\rho _{k-1}}&=\frac{\gamma _k}{\gamma _{k-1}}=\phi , k>1
\end{align*}
$$



# 四、非平稳性

* 非平稳序列计算出的样本ACF通常也能表明非平稳性。
* 通常，一次或者最多两次差分，可能伴随着对数或其它形式的转换，将完成平稳性的化简。
* 任意平稳时间序列的差分仍然是平稳的，但是过度差分会给序列带来不必要的相关性并使建模过程复杂化。
* Dickey-Fuller单位根检验。区分差分平稳性和趋势平稳性。

# 六、模拟时间序列的识别

## 关于样本acf函数的标准差边界 

参数[ci.type](https://www.rdocumentation.org/packages/stats/versions/3.5.3/topics/plot.acf): should the confidence limits assume a white noise input or for lag k an MA(k-1) input? Can be abbreviated. 

P84
### ci.type='white'
$$\frac{\pm 2}{\sqrt{n}}  \qquad(6.1.3)$$

### ci.type='ma'
$$\pm 2\sqrt{Var\left( r_k \right)}\approx \pm 2\sqrt{\frac{1}{n}\left[ 1+2\sum_{j=1}^q{r_{j}^{2}} \right]}, k>q\qquad(6.1.11)$$

| lag k=1 | lag k=2 | lag k=3 | lag k=4 |...|
| --- | --- | --- | --- |---|
|MA(0) input, or white noise $e_t$|MA(1) input|MA(2) input|MA(3) input|
| $\frac{\pm 2}{\sqrt{n}}$  | $$\pm 2\sqrt{\frac{1}{n}\left[ 1+2r_{1}^{2} \right]}$$   | $$\pm 2\sqrt{\frac{1}{n}\left[ 1+2\left( r_{1}^{2}+r_{2}^{2} \right) \right]}$$   | $$\pm 2\sqrt{\frac{1}{n}\left[ 1+2\left( r_{1}^{2}+r_{2}^{2}+r_{3}^{2} \right) \right]}$$   |...|


Note: 寻找$-2\sqrt{\frac{1}{n}\left[ 1+2\sum_{j=1}^q{r_{j}^{2}} \right]}<样本ACF(q+1),样本ACF(q+2),样本ACF(q+3)...<2\sqrt{\frac{1}{n}\left[ 1+2\sum_{j=1}^q{r_{j}^{2}} \right]}$开始最早的位置q+1

```r
# 加拿大野兔年度丰度序列
if(require(MASS)){
  print("成功载入包")
} else {
  print("不存在这个包，正在尝试安装")
  install.packages(MASS)
  if(require(MASS)){
    print("成功安装并载入包")
  } else {
    stop("安装失败")
  }
}
data(hare)
options(repr.plot.width=5, repr.plot.height=3)
BoxCox.ar(hare) # help(BoxCox.ar)
acf(hare^.5)
pacf(hare^.5)
# 这些图形建议对开方后的数据建立AR(2)模型或者AR(3)模型
```

```r
#石油价格序列
data(oil.price)
options(repr.plot.width=5, repr.plot.height=3)
plot(oil.price)
eacf(diff(log(oil.price))) #对数差分后，建议p=0,q=1的ARMA模型
res=armasubsets(y=diff(log(oil.price)),nar=7,nma=7,y.name='test',ar.method='ols')
plot(res) #第一个最优模型建议对数差分后的数据用AR(4)且只用1阶和4阶滞后项建模，第二个最优模型建议研究取对数后数据的ARIMA(1,1,0)模型

acf(as.vector(diff(log(oil.price))),xaxp=c(0,22,11)) #建议MA(1)模型
pacf(as.vector(diff(log(oil.price))),xaxp=c(0,22,11)) #建议AR(2)模型（忽略某些滞后处显著的尖峰值）
# 后续进行参数估计和诊断检验时，会进一步研究所有这些模型。
# 注意序列中有异常值。
```
